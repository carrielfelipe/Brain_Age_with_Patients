```python
from scipy.stats import mannwhitneyu

p_values = []

for column in X_AD_1.columns:
    # Realizar la prueba de Mann-Whitney U para cada caracterÃ­stica
    stat, p_value = mannwhitneyu(X_AD_1[column], X_AD_2[column])
    p_values.append((column, p_value))

# Convertir los resultados en un DataFrame
import pandas as pd
p_values_df = pd.DataFrame(p_values, columns=['CaracterÃ­stica', 'P-valor'])

# Mostrar las caracterÃ­sticas con p-valores menores a 0.05 (significancia estadÃ­stica)
significant_features_mannwhitneyu = p_values_df[p_values_df['P-valor'] < 0.05]
print("CaracterÃ­sticas significativamente diferentes entre X_AD_1 y X_AD_2:")
print(significant_features_mannwhitneyu)
```

```python
from scipy.stats import ttest_ind
import pandas as pd

p_values = []

for column in X_AD_1.columns:
    # Realizar la prueba t de Student para cada caracterÃ­stica
    stat, p_value = ttest_ind(X_AD_1[column], X_AD_2[column], equal_var=True)  
    p_values.append((column, p_value))

# Convertir los resultados en un DataFrame
p_values_df = pd.DataFrame(p_values, columns=['CaracterÃ­stica', 'P-valor'])

# Mostrar las caracterÃ­sticas con p-valores menores a 0.05 (significancia estadÃ­stica)
significant_features_ttest_ind = p_values_df[p_values_df['P-valor'] < 0.05]
print("CaracterÃ­sticas significativamente diferentes entre X_AD_1 y X_AD_2:")
print(significant_features_ttest_ind)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# AsegÃºrate de que significant_features estÃ© definido
# Filtrar las caracterÃ­sticas significativas
significant_columns = significant_features_ttest_ind['CaracterÃ­stica']

# Crear una figura para los boxplots
plt.figure(figsize=(15, 20))

# Iterar a travÃ©s de las caracterÃ­sticas significativas y crear un boxplot para cada una
for i, column in enumerate(significant_columns):
    plt.subplot(len(significant_columns), 1, i + 1)
    # Crear un DataFrame combinado con las dos muestras
    combined_df = pd.DataFrame({
        'Valor': pd.concat([X_AD_1[column], X_AD_2[column]]),
        'Grupo': ['X_AD_1'] * len(X_AD_1[column]) + ['X_AD_2'] * len(X_AD_2[column])
    })
    # Crear el boxplot horizontal
    sns.boxplot(x='Valor', y='Grupo', data=combined_df)
    plt.title(f'{column}', fontsize=16)
    plt.xlabel('Valor')
    plt.ylabel('Grupo')

# Ajustar el diseÃ±o para evitar solapamiento
plt.tight_layout()
plt.show()
```

```python
# Calcular la media de cada caracterÃ­stica en ambos conjuntos
mean_diff = X_AD_1.mean() - X_AD_2.mean()

# Ordenar las caracterÃ­sticas por la mayor diferencia absoluta en la media
mean_diff_sorted = mean_diff.abs().sort_values(ascending=False)
print("CaracterÃ­sticas con mayores diferencias en las medias:")
print(mean_diff_sorted.head(10))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import Ridge, Lasso

scaler = MinMaxScaler((0.05,0.95))
#scaler = StandardScaler()

scaler.fit(X_CN)

X_CN_scaled = scaler.transform(X_CN)
X_AD_scaled_1 = scaler.transform(X_AD_1)
X_AD_scaled_2 = scaler.transform(X_AD_2)
X_FTD_scaled = scaler.transform(X_FTD)
X_MCI_scaled = scaler.transform(X_MCI)
X_PD_scaled = scaler.transform(X_PD)

X_CN_train, X_CN_test, y_CN_train, y_CN_test = train_test_split(X_CN_scaled, y_CN, test_size=0.2, random_state=42)

##################################
#model = LinearRegression()
#model.fit(X_CN_train, y_CN_train)

#model = Ridge(alpha=1.0)
#model.fit(X_CN_train, y_CN_train)

model = Lasso(alpha=0.1)
model.fit(X_CN_train, y_CN_train)


############################3
y_pred_CN_test = model.predict(X_CN_test)

mse_CN_test = mean_squared_error(y_CN_test, y_pred_CN_test)
r2_CN_test = r2_score(y_CN_test, y_pred_CN_test)
print(f"Resultados en CN (test): MSE = {mse_CN_test}, R2 = {r2_CN_test}")


##########
y_pred_AD_1 = model.predict(X_AD_scaled_1)

mse_AD_1 = mean_squared_error(y_AD_1, y_pred_AD_1)
r2_AD_1 = r2_score(y_AD_1, y_pred_AD_1)
print(f"Resultados en AD_1: MSE = {mse_AD_1}, R2 = {r2_AD_1}")

residuals_AD_1 = y_AD_1 - y_pred_AD_1  # Calcular los residuales (errores)
print(f"Residuals en AD_1: {residuals_AD_1}")

#####
y_pred_AD_2 = model.predict(X_AD_scaled_2)

mse_AD_2 = mean_squared_error(y_AD_2, y_pred_AD_2)
r2_AD_2 = r2_score(y_AD_2, y_pred_AD_2)
print(f"Resultados en AD_2: MSE = {mse_AD_2}, R2 = {r2_AD_2}")

residuals_AD_2 = y_AD_2 - y_pred_AD_2  # Calcular los residuales (errores)
print(f"Residuals en AD_2: {residuals_AD_2}")


#######
# Para FTD
y_pred_FTD = model.predict(X_FTD_scaled)

mse_FTD = mean_squared_error(y_FTD, y_pred_FTD)
r2_FTD = r2_score(y_FTD, y_pred_FTD)
print(f"Resultados en FTD: MSE = {mse_FTD}, R2 = {r2_FTD}")

residuals_FTD = y_FTD - y_pred_FTD  # Calcular los residuales (errores)
print(f"Residuals en FTD: {residuals_FTD}")

# Para MCI
y_pred_MCI = model.predict(X_MCI_scaled)

mse_MCI = mean_squared_error(y_MCI, y_pred_MCI)
r2_MCI = r2_score(y_MCI, y_pred_MCI)
print(f"Resultados en MCI: MSE = {mse_MCI}, R2 = {r2_MCI}")

residuals_MCI = y_MCI - y_pred_MCI  # Calcular los residuales (errores)
print(f"Residuals en MCI: {residuals_MCI}")

# Para PD
y_pred_PD = model.predict(X_PD_scaled)

mse_PD = mean_squared_error(y_PD, y_pred_PD)
r2_PD = r2_score(y_PD, y_pred_PD)
print(f"Resultados en PD: MSE = {mse_PD}, R2 = {r2_PD}")

residuals_PD = y_PD - y_pred_PD  # Calcular los residuales (errores)
print(f"Residuals en PD: {residuals_PD}")

```
