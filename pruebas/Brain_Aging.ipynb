# Seleccion de caracteristicas

```python
from scipy.stats import ttest_ind
import pandas as pd

# Listado de combinaciones de dataframes
dataframes = {'CN': X_CN, 'AD': X_AD, 'MCI': X_MCI, 'FTD': X_FTD, 'PD': X_PD}
combinaciones = [(key1, key2) for key1 in dataframes.keys() for key2 in dataframes.keys() if key1 != key2]

# DataFrame para almacenar resultados
resultados = pd.DataFrame(columns=['Combinación', 'Característica', 'p-valor'])

# Realizar la prueba t para cada combinación y cada característica
for (grupo1, grupo2) in combinaciones:
    df1 = dataframes[grupo1]
    df2 = dataframes[grupo2]
    
    # Asegurarse de que ambos dataframes tengan las mismas columnas
    assert df1.columns.equals(df2.columns), "Las columnas no coinciden entre los grupos"
    
    # Prueba t para cada característica
    for columna in df1.columns:
        stat, p_valor = ttest_ind(df1[columna], df2[columna], equal_var=False)  # Prueba t de dos muestras independientes
        nueva_fila = pd.DataFrame({'Combinación': [f'{grupo1} vs {grupo2}'], 
                                   'Característica': [columna], 
                                   'p-valor': [p_valor]})
        
        # Concatenar la nueva fila al DataFrame de resultados
        resultados = pd.concat([resultados, nueva_fila], ignore_index=True)

# Filtrar las características más significativas
umbral = 0.05
caracteristicas_significativas = resultados[resultados['p-valor'] < umbral]

# Mostrar los resultados
print(caracteristicas_significativas)
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import linregress
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
import pandas as pd


X_CN = df_CN_filtrado[features]
y_CN = df_CN_filtrado["Age"]
ID_CN = df_CN_filtrado["ID-unique"]

X_AD = df_AD_filtrado[features]
y_AD = df_AD_filtrado["Age"]
ID_AD = df_AD_filtrado["ID-unique"]

X_FTD = df_FTD_filtrado[features]
y_FTD = df_FTD_filtrado["Age"]
ID_FTD = df_FTD_filtrado["ID-unique"]

X_MCI = df_MCI_filtrado[features]
y_MCI = df_MCI_filtrado["Age"]
ID_MCI = df_MCI_filtrado["ID-unique"]

X_PD = df_PD_filtrado[features]
y_PD = df_PD_filtrado["Age"]
ID_PD = df_PD_filtrado["ID-unique"]


results = { 'model': []}

results_per_fold_AD = {}
results_per_fold_CN = {}
results_per_fold_FTD = {}
results_per_fold_MCI = {}
results_per_fold_PD = {}

results_labels_df_CN = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_AD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_FTD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_MCI = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_PD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])

n_splits = 10
n_iterations = 20

for i in range(n_iterations):
    kf_CN = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_AD = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_FTD = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_MCI = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_PD = KFold(n_splits=n_splits, shuffle=True, random_state=i)

    # Dividir X_AD_scaled en 10 folds
    kf_AD_splits = list(kf_AD.split(X_AD, y_AD))
    kf_FTD_splits = list(kf_FTD.split(X_FTD, y_FTD))
    kf_MCI_splits = list(kf_MCI.split(X_MCI, y_MCI))
    kf_PD_splits = list(kf_PD.split(X_PD, y_PD))

    for fold_CN, (train_index_CN, test_index_CN) in enumerate(kf_CN.split(X_CN, y_CN)):
        X_train_kf_CN, X_test_kf_CN = X_CN.iloc[train_index_CN], X_CN.iloc[test_index_CN]
        y_train_kf_CN, y_test_kf_CN = y_CN.iloc[train_index_CN], y_CN.iloc[test_index_CN]
        id_train_kf_CN, id_test_kf_CN = ID_CN.iloc[train_index_CN], ID_CN.iloc[test_index_CN]
        
        mean_X_train_kf_CN = X_train_kf_CN.mean()
        std_X_train_kf_CN = X_train_kf_CN.std()

        X_train_kf_CN_scaled  = (X_train_kf_CN - mean_X_train_kf_CN) / std_X_train_kf_CN
        X_test_kf_CN_scaled  = (X_test_kf_CN - mean_X_train_kf_CN) / std_X_train_kf_CN
        
        # Crear y entrenar el modelo en el fold actual
        model = Lasso(alpha=0.2, max_iter=10000)  
        model.fit(X_train_kf_CN_scaled, y_train_kf_CN)

        # Predicciones para el conjunto de entrenamiento
        y_pred_train = model.predict(X_train_kf_CN_scaled)

        # Predicciones para el conjunto de prueba
        y_pred_test = model.predict(X_test_kf_CN_scaled)

        # Obtener el fold correspondiente de X_AD_scaled
        train_index_AD, test_index_AD = kf_AD_splits[fold_CN]
        X_test_kf_AD = X_AD.iloc[test_index_AD]
        X_test_kf_AD_scaled = (X_test_kf_AD - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_AD = y_AD.iloc[test_index_AD]
        id_train_kf_AD, id_test_kf_AD = ID_AD.iloc[train_index_AD], ID_AD.iloc[test_index_AD]

        # Predicciones para el fold correspondiente de AD usando el modelo entrenado en el fold CN
        y_pred_AD = model.predict(X_test_kf_AD_scaled)
        # Obtener el fold correspondiente de X_FTD_scaled
        train_index_FTD, test_index_FTD = kf_FTD_splits[fold_CN]
        X_test_kf_FTD = X_FTD.iloc[test_index_FTD]
        X_test_kf_FTD_scaled = (X_test_kf_FTD - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_FTD = y_FTD.iloc[test_index_FTD]
        id_train_kf_FTD, id_test_kf_FTD = ID_FTD.iloc[train_index_FTD], ID_FTD.iloc[test_index_FTD]

        # Predicciones para el fold correspondiente de FTD usando el modelo entrenado en el fold CN
        y_pred_FTD = model.predict(X_test_kf_FTD_scaled)

        # Obtener el fold correspondiente de X_MCI_scaled
        train_index_MCI, test_index_MCI = kf_MCI_splits[fold_CN]
        X_test_kf_MCI = X_MCI.iloc[test_index_MCI]
        X_test_kf_MCI_scaled = (X_test_kf_MCI - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_MCI = y_MCI.iloc[test_index_MCI]
        id_train_kf_MCI, id_test_kf_MCI = ID_MCI.iloc[train_index_MCI], ID_MCI.iloc[test_index_MCI]

        # Predicciones para el fold correspondiente de MCI usando el modelo entrenado en el fold CN
        y_pred_MCI = model.predict(X_test_kf_MCI_scaled)

        # Obtener el fold correspondiente de X_PD_scaled
        train_index_PD, test_index_PD = kf_PD_splits[fold_CN]
        X_test_kf_PD = X_PD.iloc[test_index_PD]
        X_test_kf_PD_scaled = (X_test_kf_PD - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_PD = y_PD.iloc[test_index_PD]
        id_train_kf_PD, id_test_kf_PD = ID_PD.iloc[train_index_PD], ID_PD.iloc[test_index_PD]

        # Predicciones para el fold correspondiente de PD usando el modelo entrenado en el fold CN
        y_pred_PD = model.predict(X_test_kf_PD_scaled)

        # Calcular los GAPs
        gap_test = y_pred_test - y_test_kf_CN
        gap_train = y_pred_train - y_train_kf_CN
        gap_AD = y_pred_AD - y_test_kf_AD
        gap_FTD = y_pred_FTD - y_test_kf_FTD
        gap_MCI = y_pred_MCI - y_test_kf_MCI
        gap_PD = y_pred_PD - y_test_kf_PD

        # Ajuste de la regresión lineal sobre el conjunto de entrenamiento
        slope, intercept, _, _, _ = linregress(y_train_kf_CN, gap_train)

        # Corrección de GAPs para el conjunto de prueba, AD, FTD, MCI y PD
        corrected_gap = gap_test - (slope * y_test_kf_CN + intercept)
        corrected_gap_AD = gap_AD - (slope * y_test_kf_AD + intercept)
        corrected_gap_FTD = gap_FTD - (slope * y_test_kf_FTD + intercept)
        corrected_gap_MCI = gap_MCI - (slope * y_test_kf_MCI + intercept)
        corrected_gap_PD = gap_PD - (slope * y_test_kf_PD + intercept)

        # Predicciones corregidas
        y_pred_corrected = y_pred_test - (slope * y_test_kf_CN + intercept)
        y_pred_corrected_AD = y_pred_AD - (slope * y_test_kf_AD + intercept)
        y_pred_corrected_FTD = y_pred_FTD - (slope * y_test_kf_FTD + intercept)
        y_pred_corrected_MCI = y_pred_MCI - (slope * y_test_kf_MCI + intercept)
        y_pred_corrected_PD = y_pred_PD - (slope * y_test_kf_PD + intercept)

        # Almacenar los resultados de predicciones y correcciones para controles (CN) en cada fold
        temp_CN_df = pd.DataFrame({
            'y_labels': y_test_kf_CN,
            'y_pred': y_pred_test,
            'y_pred_corrected': y_pred_corrected,
            'GAP': gap_test,
            'GAP_corrected': corrected_gap,
            'ID': id_test_kf_CN
        })
        results_labels_df_CN = pd.concat([results_labels_df_CN, temp_CN_df], ignore_index=True)
        results_per_fold_CN[fold_CN] = temp_CN_df.copy()

        # Almacenar los resultados de predicciones y correcciones para AD
        temp_AD_df = pd.DataFrame({
            'y_labels': y_test_kf_AD,
            'y_pred': y_pred_AD,
            'y_pred_corrected': y_pred_corrected_AD,
            'GAP': gap_AD,
            'GAP_corrected': corrected_gap_AD,
            'ID': id_test_kf_AD
        })
        results_labels_df_AD = pd.concat([results_labels_df_AD, temp_AD_df], ignore_index=True)
        results_per_fold_AD[fold_CN] = temp_AD_df.copy()

        # Almacenar los resultados de predicciones y correcciones para FTD
        temp_FTD_df = pd.DataFrame({
            'y_labels': y_test_kf_FTD,
            'y_pred': y_pred_FTD,
            'y_pred_corrected': y_pred_corrected_FTD,
            'GAP': gap_FTD,
            'GAP_corrected': corrected_gap_FTD,
            'ID': id_test_kf_FTD
        })
        results_labels_df_FTD = pd.concat([results_labels_df_FTD, temp_FTD_df], ignore_index=True)
        results_per_fold_FTD[fold_CN] = temp_FTD_df.copy()

        # Almacenar los resultados de predicciones y correcciones para MCI
        temp_MCI_df = pd.DataFrame({
            'y_labels': y_test_kf_MCI,
            'y_pred': y_pred_MCI,
            'y_pred_corrected': y_pred_corrected_MCI,
            'GAP': gap_MCI,
            'GAP_corrected': corrected_gap_MCI,
            'ID': id_test_kf_MCI
        })
        results_labels_df_MCI = pd.concat([results_labels_df_MCI, temp_MCI_df], ignore_index=True)
        results_per_fold_MCI[fold_CN] = temp_MCI_df.copy()

        # Almacenar los resultados de predicciones y correcciones para PD
        temp_PD_df = pd.DataFrame({
            'y_labels': y_test_kf_PD,
            'y_pred': y_pred_PD,
            'y_pred_corrected': y_pred_corrected_PD,
            'GAP': gap_PD,
            'GAP_corrected': corrected_gap_PD,
            'ID': id_test_kf_PD
        })
        results_labels_df_PD = pd.concat([results_labels_df_PD, temp_PD_df], ignore_index=True)
        results_per_fold_PD[fold_CN] = temp_PD_df.copy()

        # Almacenar el modelo
        results['model'].append(model)


# Promediar los valores por ID en results_labels_df_CN
results_avg_CN = results_labels_df_CN.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_AD
results_avg_AD = results_labels_df_AD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_FTD
results_avg_FTD = results_labels_df_FTD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_MCI
results_avg_MCI = results_labels_df_MCI.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_PD
results_avg_PD = results_labels_df_PD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

print("Results average for CN:")
print(results_avg_CN.head())

print("Results average for AD:")
print(results_avg_AD.head())

print("Results average for FTD:")
print(results_avg_FTD.head())

print("Results average for MCI:")
print(results_avg_MCI.head())

print("Results average for PD:")
print(results_avg_PD.head())
```

```python

from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import linregress
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet

X_CN = df_CN_filtrado[features]
y_CN = df_CN_filtrado["Age"]
ID_CN = df_CN_filtrado["ID-unique"]

X_MCI = df_MCI_filtrado[features]
y_MCI = df_MCI_filtrado["Age"]
ID_MCI = df_MCI_filtrado["ID-unique"]


results = { 'model': []}

results_per_fold_CN = {}
results_per_fold_MCI = {}

results_labels_df_CN = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_PD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])


# Combinar AD, PD y FTD en un solo conjunto
X_AD_PD_FTD = pd.concat([X_AD, X_PD, X_FTD], axis=0)
y_AD_PD_FTD = pd.concat([y_AD, y_PD, y_FTD], axis=0)
ID_AD_PD_FTD = pd.concat([ID_AD, ID_PD, ID_FTD], axis=0)

# Crear estructuras de resultados para el conjunto combinado
results_per_fold_AD_PD_FTD = {}
results_labels_df_AD_PD_FTD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])

n_splits = 10
n_iterations = 20

for i in range(n_iterations):
    kf_CN = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_MCI = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_AD_PD_FTD = KFold(n_splits=n_splits, shuffle=True, random_state=i)

    # Dividir X_MCI_scaled en 10 folds
    kf_MCI_splits = list(kf_MCI.split(X_MCI, y_MCI))
    kf_AD_PD_FTD_splits = list(kf_AD_PD_FTD.split(X_AD_PD_FTD, y_AD_PD_FTD))

    for fold_CN, (train_index_CN, test_index_CN) in enumerate(kf_CN.split(X_CN, y_CN)):
        X_train_kf_CN, X_test_kf_CN = X_CN.iloc[train_index_CN], X_CN.iloc[test_index_CN]
        y_train_kf_CN, y_test_kf_CN = y_CN.iloc[train_index_CN], y_CN.iloc[test_index_CN]
        id_train_kf_CN, id_test_kf_CN = ID_CN.iloc[train_index_CN], ID_CN.iloc[test_index_CN]
        
        mean_X_train_kf_CN = X_train_kf_CN.mean()
        std_X_train_kf_CN = X_train_kf_CN.std()

        X_train_kf_CN_scaled  = (X_train_kf_CN - mean_X_train_kf_CN) / std_X_train_kf_CN
        X_test_kf_CN_scaled  = (X_test_kf_CN - mean_X_train_kf_CN) / std_X_train_kf_CN
        
        # Crear y entrenar el modelo en el fold actual
        model = Lasso(alpha=0.2, max_iter=10000)
        model.fit(X_train_kf_CN_scaled, y_train_kf_CN)

        y_pred_train = model.predict(X_train_kf_CN_scaled)

        # Predicciones para el conjunto de prueba
        y_pred_test = model.predict(X_test_kf_CN_scaled)

        # Obtener el fold correspondiente de MCI
        train_index_MCI, test_index_MCI = kf_MCI_splits[fold_CN]
        X_test_kf_MCI = X_MCI.iloc[test_index_MCI]
        X_test_kf_MCI_scaled = (X_test_kf_MCI - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_MCI = y_MCI.iloc[test_index_MCI]
        id_train_kf_MCI, id_test_kf_MCI = ID_MCI.iloc[train_index_MCI], ID_MCI.iloc[test_index_MCI]

        # Predicciones para el fold correspondiente de MCI
        y_pred_MCI = model.predict(X_test_kf_MCI_scaled)

        # Obtener el fold correspondiente de AD_PD_FTD
        train_index_AD_PD_FTD, test_index_AD_PD_FTD = kf_AD_PD_FTD_splits[fold_CN]
        X_test_kf_AD_PD_FTD = X_AD_PD_FTD.iloc[test_index_AD_PD_FTD]
        X_test_kf_AD_PD_FTD_scaled = (X_test_kf_AD_PD_FTD - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_AD_PD_FTD = y_AD_PD_FTD.iloc[test_index_AD_PD_FTD]
        id_train_kf_AD_PD_FTD, id_test_kf_AD_PD_FTD = ID_AD_PD_FTD.iloc[train_index_AD_PD_FTD], ID_AD_PD_FTD.iloc[test_index_AD_PD_FTD]

        # Predicciones para el fold correspondiente de AD_PD_FTD
        y_pred_AD_PD_FTD = model.predict(X_test_kf_AD_PD_FTD_scaled)

        # Calcular los GAPs
        gap_test = y_pred_test - y_test_kf_CN
        gap_train = y_pred_train - y_train_kf_CN
        gap_MCI = y_pred_MCI - y_test_kf_MCI
        gap_AD_PD_FTD = y_pred_AD_PD_FTD - y_test_kf_AD_PD_FTD

        # Ajuste de la regresión lineal sobre el conjunto de entrenamiento
        slope, intercept, _, _, _ = linregress(y_train_kf_CN, gap_train)

        # Corrección de GAPs para el conjunto de prueba, MCI y AD_PD_FTD
        corrected_gap = gap_test - (slope * y_test_kf_CN + intercept)
        corrected_gap_MCI = gap_MCI - (slope * y_test_kf_MCI + intercept)
        corrected_gap_AD_PD_FTD = gap_AD_PD_FTD - (slope * y_test_kf_AD_PD_FTD + intercept)

        # Predicciones corregidas
        y_pred_corrected = y_pred_test - (slope * y_test_kf_CN + intercept)
        y_pred_corrected_MCI = y_pred_MCI - (slope * y_test_kf_MCI + intercept)
        y_pred_corrected_AD_PD_FTD = y_pred_AD_PD_FTD - (slope * y_test_kf_AD_PD_FTD + intercept)

        # Almacenar los resultados de predicciones y correcciones para CN
        temp_CN_df = pd.DataFrame({
            'y_labels': y_test_kf_CN,
            'y_pred': y_pred_test,
            'y_pred_corrected': y_pred_corrected,
            'GAP': gap_test,
            'GAP_corrected': corrected_gap,
            'ID': id_test_kf_CN
        })
        results_labels_df_CN = pd.concat([results_labels_df_CN, temp_CN_df], ignore_index=True)
        results_per_fold_CN[fold_CN] = temp_CN_df.copy()

        # Almacenar los resultados de predicciones y correcciones para MCI
        temp_MCI_df = pd.DataFrame({
            'y_labels': y_test_kf_MCI,
            'y_pred': y_pred_MCI,
            'y_pred_corrected': y_pred_corrected_MCI,
            'GAP': gap_MCI,
            'GAP_corrected': corrected_gap_MCI,
            'ID': id_test_kf_MCI
        })
        results_labels_df_MCI = pd.concat([results_labels_df_MCI, temp_MCI_df], ignore_index=True)
        results_per_fold_MCI[fold_CN] = temp_MCI_df.copy()

        # Almacenar los resultados de predicciones y correcciones para AD_PD_FTD
        temp_AD_PD_FTD_df = pd.DataFrame({
            'y_labels': y_test_kf_AD_PD_FTD,
            'y_pred': y_pred_AD_PD_FTD,
            'y_pred_corrected': y_pred_corrected_AD_PD_FTD,
            'GAP': gap_AD_PD_FTD,
            'GAP_corrected': corrected_gap_AD_PD_FTD,
            'ID': id_test_kf_AD_PD_FTD
        })
        results_labels_df_AD_PD_FTD = pd.concat([results_labels_df_AD_PD_FTD, temp_AD_PD_FTD_df], ignore_index=True)
        results_per_fold_AD_PD_FTD[fold_CN] = temp_AD_PD_FTD_df.copy()

        # Almacenar el modelo
        results['model'].append(model)

# Promediar los valores por ID en results_labels_df_CN
results_avg_CN = results_labels_df_CN.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()




#######################
# Promediar los valores por ID en results_labels_df_MCI
results_avg_MCI = results_labels_df_MCI.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_PD
results_avg_AD_PD_FTD = results_labels_df_AD_PD_FTD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

print("Results average for CN:")
print(results_avg_CN.head())

print("Results average for MCI:")
print(results_avg_MCI.head())

print("Results average for AD_PD_FTD:")
print(results_avg_AD_PD_FTD.head())


```

```python

from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import linregress
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet

X_CN = df_CN_filtrado[features]
y_CN = df_CN_filtrado["Age"]
ID_CN = df_CN_filtrado["ID-unique"]

X_AD = df_AD_filtrado[features]
y_AD = df_AD_filtrado["Age"]
ID_AD = df_AD_filtrado["ID-unique"]

X_FTD = df_FTD_filtrado[features]
y_FTD = df_FTD_filtrado["Age"]
ID_FTD = df_FTD_filtrado["ID-unique"]

X_MCI = df_MCI_filtrado[features]
y_MCI = df_MCI_filtrado["Age"]
ID_MCI = df_MCI_filtrado["ID-unique"]

X_PD = df_PD_filtrado[features]
y_PD = df_PD_filtrado["Age"]
ID_PD = df_PD_filtrado["ID-unique"]


results = { 'model': []}

results_per_fold_CN = {}
results_per_fold_MCI = {}

results_labels_df_CN = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_PD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])


# Combinar AD, PD y FTD en un solo conjunto
X_AD_PD_FTD = pd.concat([X_AD, X_PD, X_FTD], axis=0)
y_AD_PD_FTD = pd.concat([y_AD, y_PD, y_FTD], axis=0)
ID_AD_PD_FTD = pd.concat([ID_AD, ID_PD, ID_FTD], axis=0)

# Crear estructuras de resultados para el conjunto combinado
results_per_fold_AD_PD_FTD = {}
results_labels_df_AD_PD_FTD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])

n_splits = 10
n_iterations = 20

for i in range(n_iterations):
    kf_CN = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_MCI = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_AD_PD_FTD = KFold(n_splits=n_splits, shuffle=True, random_state=i)

    # Dividir X_MCI_scaled en 10 folds
    kf_MCI_splits = list(kf_MCI.split(X_MCI, y_MCI))
    kf_AD_PD_FTD_splits = list(kf_AD_PD_FTD.split(X_AD_PD_FTD, y_AD_PD_FTD))

    for fold_CN, (train_index_CN, test_index_CN) in enumerate(kf_CN.split(X_CN, y_CN)):
        X_train_kf_CN, X_test_kf_CN = X_CN.iloc[train_index_CN], X_CN.iloc[test_index_CN]
        y_train_kf_CN, y_test_kf_CN = y_CN.iloc[train_index_CN], y_CN.iloc[test_index_CN]
        id_train_kf_CN, id_test_kf_CN = ID_CN.iloc[train_index_CN], ID_CN.iloc[test_index_CN]
        
        mean_X_train_kf_CN = X_train_kf_CN.mean()
        std_X_train_kf_CN = X_train_kf_CN.std()

        X_train_kf_CN_scaled  = (X_train_kf_CN - mean_X_train_kf_CN) / std_X_train_kf_CN
        X_test_kf_CN_scaled  = (X_test_kf_CN - mean_X_train_kf_CN) / std_X_train_kf_CN
        
        # Crear y entrenar el modelo en el fold actual
        model = Lasso(alpha=0.2, max_iter=10000)
        model.fit(X_train_kf_CN_scaled, y_train_kf_CN)

        y_pred_train = model.predict(X_train_kf_CN_scaled)

        # Predicciones para el conjunto de prueba
        y_pred_test = model.predict(X_test_kf_CN_scaled)

        # Obtener el fold correspondiente de MCI
        train_index_MCI, test_index_MCI = kf_MCI_splits[fold_CN]
        X_test_kf_MCI = X_MCI.iloc[test_index_MCI]
        X_test_kf_MCI_scaled = (X_test_kf_MCI - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_MCI = y_MCI.iloc[test_index_MCI]
        id_train_kf_MCI, id_test_kf_MCI = ID_MCI.iloc[train_index_MCI], ID_MCI.iloc[test_index_MCI]

        # Predicciones para el fold correspondiente de MCI
        y_pred_MCI = model.predict(X_test_kf_MCI_scaled)

        # Obtener el fold correspondiente de AD_PD_FTD
        train_index_AD_PD_FTD, test_index_AD_PD_FTD = kf_AD_PD_FTD_splits[fold_CN]
        X_test_kf_AD_PD_FTD = X_AD_PD_FTD.iloc[test_index_AD_PD_FTD]
        X_test_kf_AD_PD_FTD_scaled = (X_test_kf_AD_PD_FTD - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_AD_PD_FTD = y_AD_PD_FTD.iloc[test_index_AD_PD_FTD]
        id_train_kf_AD_PD_FTD, id_test_kf_AD_PD_FTD = ID_AD_PD_FTD.iloc[train_index_AD_PD_FTD], ID_AD_PD_FTD.iloc[test_index_AD_PD_FTD]

        # Predicciones para el fold correspondiente de AD_PD_FTD
        y_pred_AD_PD_FTD = model.predict(X_test_kf_AD_PD_FTD_scaled)

        # Calcular los GAPs
        gap_test = y_pred_test - y_test_kf_CN
        gap_train = y_pred_train - y_train_kf_CN
        gap_MCI = y_pred_MCI - y_test_kf_MCI
        gap_AD_PD_FTD = y_pred_AD_PD_FTD - y_test_kf_AD_PD_FTD

        # Ajuste de la regresión lineal sobre el conjunto de entrenamiento
        slope, intercept, _, _, _ = linregress(y_train_kf_CN, gap_train)

        # Corrección de GAPs para el conjunto de prueba, MCI y AD_PD_FTD
        corrected_gap = gap_test - (slope * y_test_kf_CN + intercept)
        corrected_gap_MCI = gap_MCI - (slope * y_test_kf_MCI + intercept)
        corrected_gap_AD_PD_FTD = gap_AD_PD_FTD - (slope * y_test_kf_AD_PD_FTD + intercept)

        # Predicciones corregidas
        y_pred_corrected = y_pred_test - (slope * y_test_kf_CN + intercept)
        y_pred_corrected_MCI = y_pred_MCI - (slope * y_test_kf_MCI + intercept)
        y_pred_corrected_AD_PD_FTD = y_pred_AD_PD_FTD - (slope * y_test_kf_AD_PD_FTD + intercept)

        # Almacenar los resultados de predicciones y correcciones para CN
        temp_CN_df = pd.DataFrame({
            'y_labels': y_test_kf_CN,
            'y_pred': y_pred_test,
            'y_pred_corrected': y_pred_corrected,
            'GAP': gap_test,
            'GAP_corrected': corrected_gap,
            'ID': id_test_kf_CN
        })
        results_labels_df_CN = pd.concat([results_labels_df_CN, temp_CN_df], ignore_index=True)
        results_per_fold_CN[fold_CN] = temp_CN_df.copy()

        # Almacenar los resultados de predicciones y correcciones para MCI
        temp_MCI_df = pd.DataFrame({
            'y_labels': y_test_kf_MCI,
            'y_pred': y_pred_MCI,
            'y_pred_corrected': y_pred_corrected_MCI,
            'GAP': gap_MCI,
            'GAP_corrected': corrected_gap_MCI,
            'ID': id_test_kf_MCI
        })
        results_labels_df_MCI = pd.concat([results_labels_df_MCI, temp_MCI_df], ignore_index=True)
        results_per_fold_MCI[fold_CN] = temp_MCI_df.copy()

        # Almacenar los resultados de predicciones y correcciones para AD_PD_FTD
        temp_AD_PD_FTD_df = pd.DataFrame({
            'y_labels': y_test_kf_AD_PD_FTD,
            'y_pred': y_pred_AD_PD_FTD,
            'y_pred_corrected': y_pred_corrected_AD_PD_FTD,
            'GAP': gap_AD_PD_FTD,
            'GAP_corrected': corrected_gap_AD_PD_FTD,
            'ID': id_test_kf_AD_PD_FTD
        })
        results_labels_df_AD_PD_FTD = pd.concat([results_labels_df_AD_PD_FTD, temp_AD_PD_FTD_df], ignore_index=True)
        results_per_fold_AD_PD_FTD[fold_CN] = temp_AD_PD_FTD_df.copy()

        # Almacenar el modelo
        results['model'].append(model)

# Promediar los valores por ID en results_labels_df_CN
results_avg_CN = results_labels_df_CN.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()




#######################
# Promediar los valores por ID en results_labels_df_MCI
results_avg_MCI = results_labels_df_MCI.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_PD
results_avg_AD_PD_FTD = results_labels_df_AD_PD_FTD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

print("Results average for CN:")
print(results_avg_CN.head())

print("Results average for MCI:")
print(results_avg_MCI.head())

print("Results average for AD_PD_FTD:")
print(results_avg_AD_PD_FTD.head())


```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import linregress
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
import pandas as pd
import numpy as np


X_CN = df_CN_filtrado[features]
y_CN = df_CN_filtrado["Age"]
ID_CN = df_CN_filtrado["ID-unique"]

X_AD = df_AD_filtrado[features]
y_AD = df_AD_filtrado["Age"]
ID_AD = df_AD_filtrado["ID-unique"]

X_FTD = df_FTD_filtrado[features]
y_FTD = df_FTD_filtrado["Age"]
ID_FTD = df_FTD_filtrado["ID-unique"]

X_MCI = df_MCI_filtrado[features]
y_MCI = df_MCI_filtrado["Age"]
ID_MCI = df_MCI_filtrado["ID-unique"]

X_PD = df_PD_filtrado[features]
y_PD = df_PD_filtrado["Age"]
ID_PD = df_PD_filtrado["ID-unique"]


results = { 'model': []}

results_per_fold_AD = {}
results_per_fold_CN = {}
results_per_fold_FTD = {}
results_per_fold_MCI = {}
results_per_fold_PD = {}

results_labels_df_CN = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_AD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_FTD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_MCI = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_PD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])

n_splits = 10
n_iterations = 20

for i in range(n_iterations):
    kf_CN = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_AD = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_FTD = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_MCI = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_PD = KFold(n_splits=n_splits, shuffle=True, random_state=i)

    kf_CN_splits = list(kf_CN.split(X_CN, y_CN))
    kf_AD_splits = list(kf_AD.split(X_AD, y_AD))
    kf_FTD_splits = list(kf_FTD.split(X_FTD, y_FTD))
    kf_MCI_splits = list(kf_MCI.split(X_MCI, y_MCI))
    kf_PD_splits = list(kf_PD.split(X_PD, y_PD))

    for fold in range(n_splits):

        train_index_CN, test_index_CN = kf_CN_splits[fold]
        X_train_kf_CN, X_test_kf_CN = X_CN.iloc[train_index_CN], X_CN.iloc[test_index_CN]
        y_train_kf_CN, y_test_kf_CN = y_CN.iloc[train_index_CN], y_CN.iloc[test_index_CN]
        id_train_kf_CN, id_test_kf_CN = ID_CN.iloc[train_index_CN], ID_CN.iloc[test_index_CN]

        train_index_AD, test_index_AD = kf_AD_splits[fold]
        X_train_kf_AD, X_test_kf_AD = X_AD.iloc[train_index_AD], X_AD.iloc[test_index_AD]
        y_train_kf_AD, y_test_kf_AD = y_AD.iloc[train_index_AD], y_AD.iloc[test_index_AD]
        id_train_kf_AD, id_test_kf_AD = ID_AD.iloc[train_index_AD], ID_AD.iloc[test_index_AD]

        train_index_FTD, test_index_FTD = kf_FTD_splits[fold]
        X_train_kf_FTD, X_test_kf_FTD = X_FTD.iloc[train_index_FTD], X_FTD.iloc[test_index_FTD]
        y_train_kf_FTD, y_test_kf_FTD = y_FTD.iloc[train_index_FTD], y_FTD.iloc[test_index_FTD]
        id_train_kf_FTD, id_test_kf_FTD = ID_FTD.iloc[train_index_FTD], ID_FTD.iloc[test_index_FTD]

        train_index_MCI, test_index_MCI = kf_MCI_splits[fold]
        X_train_kf_MCI, X_test_kf_MCI = X_MCI.iloc[train_index_MCI], X_MCI.iloc[test_index_MCI]
        y_train_kf_MCI, y_test_kf_MCI = y_MCI.iloc[train_index_MCI], y_MCI.iloc[test_index_MCI]
        id_train_kf_MCI, id_test_kf_MCI = ID_MCI.iloc[train_index_MCI], ID_MCI.iloc[test_index_MCI]

        train_index_PD, test_index_PD = kf_PD_splits[fold]
        X_train_kf_PD, X_test_kf_PD = X_PD.iloc[train_index_PD], X_PD.iloc[test_index_PD]
        y_train_kf_PD, y_test_kf_PD = y_PD.iloc[train_index_PD], y_PD.iloc[test_index_PD]
        id_train_kf_PD, id_test_kf_PD = ID_PD.iloc[train_index_PD], ID_PD.iloc[test_index_PD]


        X_train_combined = np.concatenate([X_train_kf_CN, X_train_kf_AD, X_train_kf_FTD, X_train_kf_MCI, X_train_kf_PD])
        y_train_combined = np.concatenate([y_train_kf_CN, y_train_kf_AD, y_train_kf_FTD, y_train_kf_MCI, y_train_kf_PD])

        mean_X_train_kf = X_train_combined.mean()
        std_X_train_kf = X_train_combined.std()


        X_train_combined_scaled = (X_train_combined - mean_X_train_kf) / std_X_train_kf

        X_train_kf_CN_scaled  = (X_train_kf_CN - mean_X_train_kf) / std_X_train_kf
        X_test_kf_CN_scaled  = (X_test_kf_CN - mean_X_train_kf) / std_X_train_kf
        
        X_train_kf_AD_scaled  = (X_train_kf_AD - mean_X_train_kf) / std_X_train_kf
        X_test_kf_AD_scaled  = (X_test_kf_AD - mean_X_train_kf) / std_X_train_kf
        
        X_train_kf_MCI_scaled  = (X_train_kf_MCI - mean_X_train_kf) / std_X_train_kf
        X_test_kf_MCI_scaled  = (X_test_kf_MCI - mean_X_train_kf) / std_X_train_kf
        
        X_train_kf_FTD_scaled  = (X_train_kf_FTD - mean_X_train_kf) / std_X_train_kf
        X_test_kf_FTD_scaled  = (X_test_kf_FTD - mean_X_train_kf) / std_X_train_kf
        
        X_train_kf_PD_scaled  = (X_train_kf_PD - mean_X_train_kf) / std_X_train_kf
        X_test_kf_PD_scaled  = (X_test_kf_PD - mean_X_train_kf) / std_X_train_kf
        

        # Crear y entrenar el modelo en el fold actual
        model = Lasso(alpha=0.1, max_iter=1000)  
        model.fit(X_train_combined_scaled, y_train_combined)


        # Predicciones para el conjunto de entrenamiento
        y_pred_train = model.predict(X_train_combined_scaled)

        # Predicciones para el conjunto de prueba
        y_pred_CN_test = model.predict(X_test_kf_CN_scaled)
        y_pred_AD_test = model.predict(X_test_kf_AD_scaled)
        y_pred_FTD_test = model.predict(X_test_kf_FTD_scaled)
        y_pred_MCI_test = model.predict(X_test_kf_MCI_scaled)
        y_pred_PD_test = model.predict(X_test_kf_PD_scaled)



        # Calcular los GAPs
        
        gap_train = y_pred_train - y_train_combined

        gap_CN = y_pred_CN_test - y_test_kf_CN
        gap_AD = y_pred_AD_test - y_test_kf_AD
        gap_FTD = y_pred_FTD_test - y_test_kf_FTD
        gap_MCI = y_pred_MCI_test - y_test_kf_MCI
        gap_PD = y_pred_PD_test - y_test_kf_PD

        slope, intercept, _, _, _ = linregress(y_train_combined, gap_train)


        # Corrección de GAPs para el conjunto de prueba, AD, FTD, MCI y PD
        corrected_gap_CN = gap_CN - (slope * y_test_kf_CN + intercept)
        corrected_gap_AD = gap_AD - (slope * y_test_kf_AD + intercept)
        corrected_gap_FTD = gap_FTD - (slope * y_test_kf_FTD + intercept)
        corrected_gap_MCI = gap_MCI - (slope * y_test_kf_MCI + intercept)
        corrected_gap_PD = gap_PD - (slope * y_test_kf_PD + intercept)

        # Predicciones corregidas
        y_pred_corrected_CN = y_pred_CN_test - (slope * y_test_kf_CN + intercept)
        y_pred_corrected_AD = y_pred_AD_test - (slope * y_test_kf_AD + intercept)
        y_pred_corrected_FTD = y_pred_FTD_test - (slope * y_test_kf_FTD + intercept)
        y_pred_corrected_MCI = y_pred_MCI_test - (slope * y_test_kf_MCI + intercept)
        y_pred_corrected_PD = y_pred_PD_test - (slope * y_test_kf_PD + intercept)

        # Almacenar los resultados de predicciones y correcciones para controles (CN) en cada fold
        temp_CN_df = pd.DataFrame({
            'y_labels': y_test_kf_CN,
            'y_pred': y_pred_CN_test,
            'y_pred_corrected': y_pred_corrected_CN,
            'GAP': gap_CN,
            'GAP_corrected': corrected_gap_CN,
            'ID': id_test_kf_CN
        })
        results_labels_df_CN = pd.concat([results_labels_df_CN, temp_CN_df], ignore_index=True)
        results_per_fold_CN[fold_CN] = temp_CN_df.copy()

        # Almacenar los resultados de predicciones y correcciones para AD
        temp_AD_df = pd.DataFrame({
            'y_labels': y_test_kf_AD,
            'y_pred': y_pred_AD_test,
            'y_pred_corrected': y_pred_corrected_AD,
            'GAP': gap_AD,
            'GAP_corrected': corrected_gap_AD,
            'ID': id_test_kf_AD
        })
        results_labels_df_AD = pd.concat([results_labels_df_AD, temp_AD_df], ignore_index=True)
        results_per_fold_AD[fold_CN] = temp_AD_df.copy()

        # Almacenar los resultados de predicciones y correcciones para FTD
        temp_FTD_df = pd.DataFrame({
            'y_labels': y_test_kf_FTD,
            'y_pred': y_pred_FTD_test,
            'y_pred_corrected': y_pred_corrected_FTD,
            'GAP': gap_FTD,
            'GAP_corrected': corrected_gap_FTD,
            'ID': id_test_kf_FTD
        })
        results_labels_df_FTD = pd.concat([results_labels_df_FTD, temp_FTD_df], ignore_index=True)
        results_per_fold_FTD[fold_CN] = temp_FTD_df.copy()

        # Almacenar los resultados de predicciones y correcciones para MCI
        temp_MCI_df = pd.DataFrame({
            'y_labels': y_test_kf_MCI,
            'y_pred': y_pred_MCI_test,
            'y_pred_corrected': y_pred_corrected_MCI,
            'GAP': gap_MCI,
            'GAP_corrected': corrected_gap_MCI,
            'ID': id_test_kf_MCI
        })
        results_labels_df_MCI = pd.concat([results_labels_df_MCI, temp_MCI_df], ignore_index=True)
        results_per_fold_MCI[fold_CN] = temp_MCI_df.copy()

        # Almacenar los resultados de predicciones y correcciones para PD
        temp_PD_df = pd.DataFrame({
            'y_labels': y_test_kf_PD,
            'y_pred': y_pred_PD_test,
            'y_pred_corrected': y_pred_corrected_PD,
            'GAP': gap_PD,
            'GAP_corrected': corrected_gap_PD,
            'ID': id_test_kf_PD
        })
        results_labels_df_PD = pd.concat([results_labels_df_PD, temp_PD_df], ignore_index=True)
        results_per_fold_PD[fold_CN] = temp_PD_df.copy()

        # Almacenar el modelo
        results['model'].append(model)


# Promediar los valores por ID en results_labels_df_CN
results_avg_CN = results_labels_df_CN.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_AD
results_avg_AD = results_labels_df_AD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_FTD
results_avg_FTD = results_labels_df_FTD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_MCI
results_avg_MCI = results_labels_df_MCI.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_PD
results_avg_PD = results_labels_df_PD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

print("Results average for CN:")
print(results_avg_CN.head())

print("Results average for AD:")
print(results_avg_AD.head())

print("Results average for FTD:")
print(results_avg_FTD.head())

print("Results average for MCI:")
print(results_avg_MCI.head())

print("Results average for PD:")
print(results_avg_PD.head())
```

```python
for fold in range(n_splits):

    kf_CN = KFold(n_splits=n_splits, shuffle=True, random_state=fold)
    kf_AD = KFold(n_splits=n_splits, shuffle=True, random_state=fold)
    kf_FTD = KFold(n_splits=n_splits, shuffle=True, random_state=fold)
    kf_MCI = KFold(n_splits=n_splits, shuffle=True, random_state=fold)
    kf_PD = KFold(n_splits=n_splits, shuffle=True, random_state=fold)

    kf_CN_splits = list(kf_CN.split(X_AD, y_AD))
    kf_AD_splits = list(kf_AD.split(X_AD, y_AD))
    kf_FTD_splits = list(kf_FTD.split(X_FTD, y_FTD))
    kf_MCI_splits = list(kf_MCI.split(X_MCI, y_MCI))
    kf_PD_splits = list(kf_PD.split(X_PD, y_PD))

    train_index_CN, test_index_CN = kf_CN_splits[fold]
    X_train_kf_CN, X_test_kf_CN = X_CN.iloc[train_index_CN], X_CN.iloc[test_index_CN]
    y_train_kf_CN, y_test_kf_CN = y_CN.iloc[train_index_CN], y_CN.iloc[test_index_CN]
    id_train_kf_CN, id_test_kf_CN = ID_CN.iloc[train_index_CN], ID_CN.iloc[test_index_CN]

    train_index_AD, test_index_AD = kf_AD_splits[fold]
    X_train_kf_AD, X_test_kf_AD = X_AD.iloc[train_index_AD], X_AD.iloc[test_index_AD]
    y_train_kf_AD, y_test_kf_AD = y_AD.iloc[train_index_AD], y_AD.iloc[test_index_AD]
    id_train_kf_AD, id_test_kf_AD = ID_AD.iloc[train_index_AD], ID_AD.iloc[test_index_AD]

    train_index_FTD, test_index_FTD = kf_FTD_splits[fold]
    X_train_kf_FTD, X_test_kf_FTD = X_FTD.iloc[train_index_FTD], X_FTD.iloc[test_index_FTD]
    y_train_kf_FTD, y_test_kf_FTD = y_FTD.iloc[train_index_FTD], y_FTD.iloc[test_index_FTD]
    id_train_kf_FTD, id_test_kf_FTD = ID_FTD.iloc[train_index_FTD], ID_FTD.iloc[test_index_FTD]

    train_index_MCI, test_index_MCI = kf_MCI_splits[fold]
    X_train_kf_MCI, X_test_kf_MCI = X_MCI.iloc[train_index_MCI], X_MCI.iloc[test_index_MCI]
    y_train_kf_MCI, y_test_kf_MCI = y_MCI.iloc[train_index_MCI], y_MCI.iloc[test_index_MCI]
    id_train_kf_MCI, id_test_kf_MCI = ID_MCI.iloc[train_index_MCI], ID_MCI.iloc[test_index_MCI]

    train_index_PD, test_index_PD = kf_PD_splits[fold]
    X_train_kf_PD, X_test_kf_PD = X_PD.iloc[train_index_PD], X_PD.iloc[test_index_PD]
    y_train_kf_PD, y_test_kf_PD = y_PD.iloc[train_index_PD], y_PD.iloc[test_index_PD]
    id_train_kf_PD, id_test_kf_PD = ID_PD.iloc[train_index_PD], ID_PD.iloc[test_index_PD]


    X_train_combined = np.concatenate([X_train_kf_CN, X_train_kf_AD, X_train_kf_FTD, X_train_kf_MCI, X_train_kf_PD])
    y_train_combined = np.concatenate([y_train_kf_CN, y_train_kf_AD, y_train_kf_FTD, y_train_kf_MCI, y_train_kf_PD])

    mean_X_train_kf = X_train_combined.mean()
    std_X_train_kf = X_train_combined.std()


    X_train_combined_scaled = (X_train_combined - mean_X_train_kf) / std_X_train_kf

    X_train_kf_CN_scaled  = (X_train_kf_CN - mean_X_train_kf) / std_X_train_kf
    X_test_kf_CN_scaled  = (X_test_kf_CN - mean_X_train_kf) / std_X_train_kf
    
    X_train_kf_AD_scaled  = (X_train_kf_AD - mean_X_train_kf) / std_X_train_kf
    X_test_kf_AD_scaled  = (X_test_kf_AD - mean_X_train_kf) / std_X_train_kf
    
    X_train_kf_MCI_scaled  = (X_train_kf_MCI - mean_X_train_kf) / std_X_train_kf
    X_test_kf_MCI_scaled  = (X_test_kf_MCI - mean_X_train_kf) / std_X_train_kf
    
    X_train_kf_FTD_scaled  = (X_train_kf_FTD - mean_X_train_kf) / std_X_train_kf
    X_test_kf_FTD_scaled  = (X_test_kf_FTD - mean_X_train_kf) / std_X_train_kf
    
    X_train_kf_PD_scaled  = (X_train_kf_PD - mean_X_train_kf) / std_X_train_kf
    X_test_kf_PD_scaled  = (X_test_kf_PD - mean_X_train_kf) / std_X_train_kf
    

    # Crear y entrenar el modelo en el fold actual
    model = Lasso(alpha=0.01, max_iter=10000)  
    model.fit(X_train_kf_CN_scaled, y_train_kf_CN)


    # Predicciones para el conjunto de entrenamiento
    y_pred_train = model.predict(X_train_combined_scaled)

    # Predicciones para el conjunto de prueba
    y_pred_CN_test = model.predict(X_test_kf_CN_scaled)
    y_pred_AD_test = model.predict(X_test_kf_AD_scaled)
    y_pred_FTD_test = model.predict(X_test_kf_FTD_scaled)
    y_pred_MCI_test = model.predict(X_test_kf_MCI_scaled)
    y_pred_PD_test = model.predict(X_test_kf_PD_scaled)



    # Calcular los GAPs
    
    gap_train = y_pred_train - y_train_combined

    gap_CN = y_pred_CN_test - y_test_kf_CN
    gap_AD = y_pred_AD_test - y_test_kf_AD
    gap_FTD = y_pred_FTD_test - y_test_kf_FTD
    gap_MCI = y_pred_MCI_test - y_test_kf_MCI
    gap_PD = y_pred_PD_test - y_test_kf_PD

    slope, intercept, _, _, _ = linregress(y_train_combined, gap_train)


    # Corrección de GAPs para el conjunto de prueba, AD, FTD, MCI y PD
    corrected_gap_CN = gap_CN - (slope * y_test_kf_CN + intercept)
    corrected_gap_AD = gap_AD - (slope * y_test_kf_AD + intercept)
    corrected_gap_FTD = gap_FTD - (slope * y_test_kf_FTD + intercept)
    corrected_gap_MCI = gap_MCI - (slope * y_test_kf_MCI + intercept)
    corrected_gap_PD = gap_PD - (slope * y_test_kf_PD + intercept)

    # Predicciones corregidas
    y_pred_corrected_CN = y_pred_CN_test - (slope * y_test_kf_CN + intercept)
    y_pred_corrected_AD = y_pred_AD_test - (slope * y_test_kf_AD + intercept)
    y_pred_corrected_FTD = y_pred_FTD_test - (slope * y_test_kf_FTD + intercept)
    y_pred_corrected_MCI = y_pred_MCI_test - (slope * y_test_kf_MCI + intercept)
    y_pred_corrected_PD = y_pred_PD_test - (slope * y_test_kf_PD + intercept)

    # Almacenar los resultados de predicciones y correcciones para controles (CN) en cada fold
    temp_CN_df = pd.DataFrame({
        'y_labels': y_test_kf_CN,
        'y_pred': y_pred_CN_test,
        'y_pred_corrected': y_pred_corrected_CN,
        'GAP': gap_CN,
        'GAP_corrected': corrected_gap_CN,
        'ID': id_test_kf_CN
    })
    results_labels_df_CN = pd.concat([results_labels_df_CN, temp_CN_df], ignore_index=True)
    results_per_fold_CN[fold_CN] = temp_CN_df.copy()

    # Almacenar los resultados de predicciones y correcciones para AD
    temp_AD_df = pd.DataFrame({
        'y_labels': y_test_kf_AD,
        'y_pred': y_pred_AD_test,
        'y_pred_corrected': y_pred_corrected_AD,
        'GAP': gap_AD,
        'GAP_corrected': corrected_gap_AD,
        'ID': id_test_kf_AD
    })
    results_labels_df_AD = pd.concat([results_labels_df_AD, temp_AD_df], ignore_index=True)
    results_per_fold_AD[fold_CN] = temp_AD_df.copy()

    # Almacenar los resultados de predicciones y correcciones para FTD
    temp_FTD_df = pd.DataFrame({
        'y_labels': y_test_kf_FTD,
        'y_pred': y_pred_FTD_test,
        'y_pred_corrected': y_pred_corrected_FTD,
        'GAP': gap_FTD,
        'GAP_corrected': corrected_gap_FTD,
        'ID': id_test_kf_FTD
    })
    results_labels_df_FTD = pd.concat([results_labels_df_FTD, temp_FTD_df], ignore_index=True)
    results_per_fold_FTD[fold_CN] = temp_FTD_df.copy()

    # Almacenar los resultados de predicciones y correcciones para MCI
    temp_MCI_df = pd.DataFrame({
        'y_labels': y_test_kf_MCI,
        'y_pred': y_pred_MCI_test,
        'y_pred_corrected': y_pred_corrected_MCI,
        'GAP': gap_MCI,
        'GAP_corrected': corrected_gap_MCI,
        'ID': id_test_kf_MCI
    })
    results_labels_df_MCI = pd.concat([results_labels_df_MCI, temp_MCI_df], ignore_index=True)
    results_per_fold_MCI[fold_CN] = temp_MCI_df.copy()

    # Almacenar los resultados de predicciones y correcciones para PD
    temp_PD_df = pd.DataFrame({
        'y_labels': y_test_kf_PD,
        'y_pred': y_pred_PD_test,
        'y_pred_corrected': y_pred_corrected_PD,
        'GAP': gap_PD,
        'GAP_corrected': corrected_gap_PD,
        'ID': id_test_kf_PD
    })
    results_labels_df_PD = pd.concat([results_labels_df_PD, temp_PD_df], ignore_index=True)
    results_per_fold_PD[fold_CN] = temp_PD_df.copy()

    # Almacenar el modelo
    results['model'].append(model)
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import linregress
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
import pandas as pd


X_CN = df_CN_filtrado[features]
y_CN = df_CN_filtrado["Age"]
ID_CN = df_CN_filtrado["ID-unique"]

X_AD = df_AD_filtrado[features]
y_AD = df_AD_filtrado["Age"]
ID_AD = df_AD_filtrado["ID-unique"]

X_FTD = df_FTD_filtrado[features]
y_FTD = df_FTD_filtrado["Age"]
ID_FTD = df_FTD_filtrado["ID-unique"]

X_MCI = df_MCI_filtrado[features]
y_MCI = df_MCI_filtrado["Age"]
ID_MCI = df_MCI_filtrado["ID-unique"]

X_PD = df_PD_filtrado[features]
y_PD = df_PD_filtrado["Age"]
ID_PD = df_PD_filtrado["ID-unique"]


results = { 'model': []}

results_per_fold_AD = {}
results_per_fold_CN = {}
results_per_fold_FTD = {}
results_per_fold_MCI = {}
results_per_fold_PD = {}

results_labels_df_CN = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_AD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_FTD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_MCI = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_PD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])

n_splits = 10
n_iterations = 20

for i in range(n_iterations):
    kf_CN = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_AD = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_FTD = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_MCI = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_PD = KFold(n_splits=n_splits, shuffle=True, random_state=i)

    # Dividir X_AD_scaled en 10 folds
    kf_AD_splits = list(kf_AD.split(X_AD, y_AD))
    kf_FTD_splits = list(kf_FTD.split(X_FTD, y_FTD))
    kf_MCI_splits = list(kf_MCI.split(X_MCI, y_MCI))
    kf_PD_splits = list(kf_PD.split(X_PD, y_PD))

    for fold_CN, (train_index_CN, test_index_CN) in enumerate(kf_CN.split(X_CN, y_CN)):
        X_train_kf_CN, X_test_kf_CN = X_CN.iloc[train_index_CN], X_CN.iloc[test_index_CN]
        y_train_kf_CN, y_test_kf_CN = y_CN.iloc[train_index_CN], y_CN.iloc[test_index_CN]
        id_train_kf_CN, id_test_kf_CN = ID_CN.iloc[train_index_CN], ID_CN.iloc[test_index_CN]
        
        mean_X_train_kf_CN = X_train_kf_CN.mean()
        std_X_train_kf_CN = X_train_kf_CN.std()

        X_train_kf_CN_scaled  = (X_train_kf_CN - mean_X_train_kf_CN) / std_X_train_kf_CN
        X_test_kf_CN_scaled  = (X_test_kf_CN - mean_X_train_kf_CN) / std_X_train_kf_CN
        
        # Crear y entrenar el modelo en el fold actual
        model = Lasso(alpha=0.2, max_iter=10000)  
        model.fit(X_train_kf_CN_scaled, y_train_kf_CN)

        # Predicciones para el conjunto de entrenamiento
        y_pred_train = model.predict(X_train_kf_CN_scaled)

        # Predicciones para el conjunto de prueba
        y_pred_test = model.predict(X_test_kf_CN_scaled)

        # Obtener el fold correspondiente de X_AD_scaled
        train_index_AD, test_index_AD = kf_AD_splits[fold_CN]
        X_test_kf_AD = X_AD.iloc[test_index_AD]
        X_test_kf_AD_scaled = (X_test_kf_AD - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_AD = y_AD.iloc[test_index_AD]
        id_train_kf_AD, id_test_kf_AD = ID_AD.iloc[train_index_AD], ID_AD.iloc[test_index_AD]

        # Predicciones para el fold correspondiente de AD usando el modelo entrenado en el fold CN
        y_pred_AD = model.predict(X_test_kf_AD_scaled)
        # Obtener el fold correspondiente de X_FTD_scaled
        train_index_FTD, test_index_FTD = kf_FTD_splits[fold_CN]
        X_test_kf_FTD = X_FTD.iloc[test_index_FTD]
        X_test_kf_FTD_scaled = (X_test_kf_FTD - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_FTD = y_FTD.iloc[test_index_FTD]
        id_train_kf_FTD, id_test_kf_FTD = ID_FTD.iloc[train_index_FTD], ID_FTD.iloc[test_index_FTD]

        # Predicciones para el fold correspondiente de FTD usando el modelo entrenado en el fold CN
        y_pred_FTD = model.predict(X_test_kf_FTD_scaled)

        # Obtener el fold correspondiente de X_MCI_scaled
        train_index_MCI, test_index_MCI = kf_MCI_splits[fold_CN]
        X_test_kf_MCI = X_MCI.iloc[test_index_MCI]
        X_test_kf_MCI_scaled = (X_test_kf_MCI - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_MCI = y_MCI.iloc[test_index_MCI]
        id_train_kf_MCI, id_test_kf_MCI = ID_MCI.iloc[train_index_MCI], ID_MCI.iloc[test_index_MCI]

        # Predicciones para el fold correspondiente de MCI usando el modelo entrenado en el fold CN
        y_pred_MCI = model.predict(X_test_kf_MCI_scaled)

        # Obtener el fold correspondiente de X_PD_scaled
        train_index_PD, test_index_PD = kf_PD_splits[fold_CN]
        X_test_kf_PD = X_PD.iloc[test_index_PD]
        X_test_kf_PD_scaled = (X_test_kf_PD - mean_X_train_kf_CN) / std_X_train_kf_CN
        y_test_kf_PD = y_PD.iloc[test_index_PD]
        id_train_kf_PD, id_test_kf_PD = ID_PD.iloc[train_index_PD], ID_PD.iloc[test_index_PD]

        # Predicciones para el fold correspondiente de PD usando el modelo entrenado en el fold CN
        y_pred_PD = model.predict(X_test_kf_PD_scaled)

        # Calcular los GAPs
        gap_test = y_pred_test - y_test_kf_CN
        gap_train = y_pred_train - y_train_kf_CN
        gap_AD = y_pred_AD - y_test_kf_AD
        gap_FTD = y_pred_FTD - y_test_kf_FTD
        gap_MCI = y_pred_MCI - y_test_kf_MCI
        gap_PD = y_pred_PD - y_test_kf_PD

        # Ajuste de la regresión lineal sobre el conjunto de entrenamiento
        slope, intercept, _, _, _ = linregress(y_train_kf_CN, gap_train)

        # Corrección de GAPs para el conjunto de prueba, AD, FTD, MCI y PD
        corrected_gap = gap_test - (slope * y_test_kf_CN + intercept)
        corrected_gap_AD = gap_AD - (slope * y_test_kf_AD + intercept)
        corrected_gap_FTD = gap_FTD - (slope * y_test_kf_FTD + intercept)
        corrected_gap_MCI = gap_MCI - (slope * y_test_kf_MCI + intercept)
        corrected_gap_PD = gap_PD - (slope * y_test_kf_PD + intercept)

        # Predicciones corregidas
        y_pred_corrected = y_pred_test - (slope * y_test_kf_CN + intercept)
        y_pred_corrected_AD = y_pred_AD - (slope * y_test_kf_AD + intercept)
        y_pred_corrected_FTD = y_pred_FTD - (slope * y_test_kf_FTD + intercept)
        y_pred_corrected_MCI = y_pred_MCI - (slope * y_test_kf_MCI + intercept)
        y_pred_corrected_PD = y_pred_PD - (slope * y_test_kf_PD + intercept)

        # Almacenar los resultados de predicciones y correcciones para controles (CN) en cada fold
        temp_CN_df = pd.DataFrame({
            'y_labels': y_test_kf_CN,
            'y_pred': y_pred_test,
            'y_pred_corrected': y_pred_corrected,
            'GAP': gap_test,
            'GAP_corrected': corrected_gap,
            'ID': id_test_kf_CN
        })
        results_labels_df_CN = pd.concat([results_labels_df_CN, temp_CN_df], ignore_index=True)
        results_per_fold_CN[fold_CN] = temp_CN_df.copy()

        # Almacenar los resultados de predicciones y correcciones para AD
        temp_AD_df = pd.DataFrame({
            'y_labels': y_test_kf_AD,
            'y_pred': y_pred_AD,
            'y_pred_corrected': y_pred_corrected_AD,
            'GAP': gap_AD,
            'GAP_corrected': corrected_gap_AD,
            'ID': id_test_kf_AD
        })
        results_labels_df_AD = pd.concat([results_labels_df_AD, temp_AD_df], ignore_index=True)
        results_per_fold_AD[fold_CN] = temp_AD_df.copy()

        # Almacenar los resultados de predicciones y correcciones para FTD
        temp_FTD_df = pd.DataFrame({
            'y_labels': y_test_kf_FTD,
            'y_pred': y_pred_FTD,
            'y_pred_corrected': y_pred_corrected_FTD,
            'GAP': gap_FTD,
            'GAP_corrected': corrected_gap_FTD,
            'ID': id_test_kf_FTD
        })
        results_labels_df_FTD = pd.concat([results_labels_df_FTD, temp_FTD_df], ignore_index=True)
        results_per_fold_FTD[fold_CN] = temp_FTD_df.copy()

        # Almacenar los resultados de predicciones y correcciones para MCI
        temp_MCI_df = pd.DataFrame({
            'y_labels': y_test_kf_MCI,
            'y_pred': y_pred_MCI,
            'y_pred_corrected': y_pred_corrected_MCI,
            'GAP': gap_MCI,
            'GAP_corrected': corrected_gap_MCI,
            'ID': id_test_kf_MCI
        })
        results_labels_df_MCI = pd.concat([results_labels_df_MCI, temp_MCI_df], ignore_index=True)
        results_per_fold_MCI[fold_CN] = temp_MCI_df.copy()

        # Almacenar los resultados de predicciones y correcciones para PD
        temp_PD_df = pd.DataFrame({
            'y_labels': y_test_kf_PD,
            'y_pred': y_pred_PD,
            'y_pred_corrected': y_pred_corrected_PD,
            'GAP': gap_PD,
            'GAP_corrected': corrected_gap_PD,
            'ID': id_test_kf_PD
        })
        results_labels_df_PD = pd.concat([results_labels_df_PD, temp_PD_df], ignore_index=True)
        results_per_fold_PD[fold_CN] = temp_PD_df.copy()

        # Almacenar el modelo
        results['model'].append(model)


# Promediar los valores por ID en results_labels_df_CN
results_avg_CN = results_labels_df_CN.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_AD
results_avg_AD = results_labels_df_AD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_FTD
results_avg_FTD = results_labels_df_FTD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_MCI
results_avg_MCI = results_labels_df_MCI.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_PD
results_avg_PD = results_labels_df_PD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

print("Results average for CN:")
print(results_avg_CN.head())

print("Results average for AD:")
print(results_avg_AD.head())

print("Results average for FTD:")
print(results_avg_FTD.head())

print("Results average for MCI:")
print(results_avg_MCI.head())

print("Results average for PD:")
print(results_avg_PD.head())
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import linregress
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
import pandas as pd
import numpy as np


X_CN = df_CN_filtrado[features]
y_CN = df_CN_filtrado["Age"]
ID_CN = df_CN_filtrado["ID-unique"]

X_AD = df_AD_filtrado[features]
y_AD = df_AD_filtrado["Age"]
ID_AD = df_AD_filtrado["ID-unique"]

X_FTD = df_FTD_filtrado[features]
y_FTD = df_FTD_filtrado["Age"]
ID_FTD = df_FTD_filtrado["ID-unique"]

X_MCI = df_MCI_filtrado[features]
y_MCI = df_MCI_filtrado["Age"]
ID_MCI = df_MCI_filtrado["ID-unique"]

X_PD = df_PD_filtrado[features]
y_PD = df_PD_filtrado["Age"]
ID_PD = df_PD_filtrado["ID-unique"]


results = { 'model': []}

results_per_fold_AD = {}
results_per_fold_CN = {}
results_per_fold_FTD = {}
results_per_fold_MCI = {}
results_per_fold_PD = {}

results_labels_df_CN = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_AD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_FTD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_MCI = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
results_labels_df_PD = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])

n_splits = 10
n_iterations = 30

for i in range(n_iterations):
    kf_CN = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_AD = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_FTD = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_MCI = KFold(n_splits=n_splits, shuffle=True, random_state=i)
    kf_PD = KFold(n_splits=n_splits, shuffle=True, random_state=i)

    kf_CN_splits = list(kf_CN.split(X_CN, y_CN))
    kf_AD_splits = list(kf_AD.split(X_AD, y_AD))
    kf_FTD_splits = list(kf_FTD.split(X_FTD, y_FTD))
    kf_MCI_splits = list(kf_MCI.split(X_MCI, y_MCI))
    kf_PD_splits = list(kf_PD.split(X_PD, y_PD))

    for fold in range(n_splits):

        train_index_CN, test_index_CN = kf_CN_splits[fold]
        X_train_kf_CN, X_test_kf_CN = X_CN.iloc[train_index_CN], X_CN.iloc[test_index_CN]
        y_train_kf_CN, y_test_kf_CN = y_CN.iloc[train_index_CN], y_CN.iloc[test_index_CN]
        id_train_kf_CN, id_test_kf_CN = ID_CN.iloc[train_index_CN], ID_CN.iloc[test_index_CN]

        train_index_AD, test_index_AD = kf_AD_splits[fold]
        X_train_kf_AD, X_test_kf_AD = X_AD.iloc[train_index_AD], X_AD.iloc[test_index_AD]
        y_train_kf_AD, y_test_kf_AD = y_AD.iloc[train_index_AD], y_AD.iloc[test_index_AD]
        id_train_kf_AD, id_test_kf_AD = ID_AD.iloc[train_index_AD], ID_AD.iloc[test_index_AD]

        train_index_FTD, test_index_FTD = kf_FTD_splits[fold]
        X_train_kf_FTD, X_test_kf_FTD = X_FTD.iloc[train_index_FTD], X_FTD.iloc[test_index_FTD]
        y_train_kf_FTD, y_test_kf_FTD = y_FTD.iloc[train_index_FTD], y_FTD.iloc[test_index_FTD]
        id_train_kf_FTD, id_test_kf_FTD = ID_FTD.iloc[train_index_FTD], ID_FTD.iloc[test_index_FTD]

        train_index_MCI, test_index_MCI = kf_MCI_splits[fold]
        X_train_kf_MCI, X_test_kf_MCI = X_MCI.iloc[train_index_MCI], X_MCI.iloc[test_index_MCI]
        y_train_kf_MCI, y_test_kf_MCI = y_MCI.iloc[train_index_MCI], y_MCI.iloc[test_index_MCI]
        id_train_kf_MCI, id_test_kf_MCI = ID_MCI.iloc[train_index_MCI], ID_MCI.iloc[test_index_MCI]

        train_index_PD, test_index_PD = kf_PD_splits[fold]
        X_train_kf_PD, X_test_kf_PD = X_PD.iloc[train_index_PD], X_PD.iloc[test_index_PD]
        y_train_kf_PD, y_test_kf_PD = y_PD.iloc[train_index_PD], y_PD.iloc[test_index_PD]
        id_train_kf_PD, id_test_kf_PD = ID_PD.iloc[train_index_PD], ID_PD.iloc[test_index_PD]
        
        mean_X_train_kf = X_train_kf_CN.mean()
        std_X_train_kf = X_train_kf_CN.std()

        X_train_kf_CN_scaled  = (X_train_kf_CN - mean_X_train_kf) / std_X_train_kf
        X_test_kf_CN_scaled  = (X_test_kf_CN - mean_X_train_kf) / std_X_train_kf
        
        X_train_kf_AD_scaled  = (X_train_kf_AD - mean_X_train_kf) / std_X_train_kf
        X_test_kf_AD_scaled  = (X_test_kf_AD - mean_X_train_kf) / std_X_train_kf
        
        X_train_kf_MCI_scaled  = (X_train_kf_MCI - mean_X_train_kf) / std_X_train_kf
        X_test_kf_MCI_scaled  = (X_test_kf_MCI - mean_X_train_kf) / std_X_train_kf
        
        X_train_kf_FTD_scaled  = (X_train_kf_FTD - mean_X_train_kf) / std_X_train_kf
        X_test_kf_FTD_scaled  = (X_test_kf_FTD - mean_X_train_kf) / std_X_train_kf
        
        X_train_kf_PD_scaled  = (X_train_kf_PD - mean_X_train_kf) / std_X_train_kf
        X_test_kf_PD_scaled  = (X_test_kf_PD - mean_X_train_kf) / std_X_train_kf
        

        # Crear y entrenar el modelo en el fold actual
        model = Lasso(alpha=0.2, max_iter=10000)  
        model.fit(X_train_kf_CN_scaled, y_train_kf_CN)


        # Predicciones para el conjunto de entrenamiento
        y_pred_train = model.predict(X_train_kf_CN_scaled)

        # Predicciones para el conjunto de prueba
        y_pred_CN_test = model.predict(X_test_kf_CN_scaled)
        y_pred_AD_test = model.predict(X_test_kf_AD_scaled)
        y_pred_FTD_test = model.predict(X_test_kf_FTD_scaled)
        y_pred_MCI_test = model.predict(X_test_kf_MCI_scaled)
        y_pred_PD_test = model.predict(X_test_kf_PD_scaled)



        # Calcular los GAPs
        
        gap_train = y_pred_train - y_train_kf_CN

        gap_CN = y_pred_CN_test - y_test_kf_CN
        gap_AD = y_pred_AD_test - y_test_kf_AD
        gap_FTD = y_pred_FTD_test - y_test_kf_FTD
        gap_MCI = y_pred_MCI_test - y_test_kf_MCI
        gap_PD = y_pred_PD_test - y_test_kf_PD

        slope, intercept, _, _, _ = linregress(y_train_kf_CN, gap_train)


        # Corrección de GAPs para el conjunto de prueba, AD, FTD, MCI y PD
        corrected_gap_CN = gap_CN - (slope * y_test_kf_CN + intercept)
        corrected_gap_AD = gap_AD - (slope * y_test_kf_AD + intercept)
        corrected_gap_FTD = gap_FTD - (slope * y_test_kf_FTD + intercept)
        corrected_gap_MCI = gap_MCI - (slope * y_test_kf_MCI + intercept)
        corrected_gap_PD = gap_PD - (slope * y_test_kf_PD + intercept)

        # Predicciones corregidas
        y_pred_corrected_CN = y_pred_CN_test - (slope * y_test_kf_CN + intercept)
        y_pred_corrected_AD = y_pred_AD_test - (slope * y_test_kf_AD + intercept)
        y_pred_corrected_FTD = y_pred_FTD_test - (slope * y_test_kf_FTD + intercept)
        y_pred_corrected_MCI = y_pred_MCI_test - (slope * y_test_kf_MCI + intercept)
        y_pred_corrected_PD = y_pred_PD_test - (slope * y_test_kf_PD + intercept)

        # Almacenar los resultados de predicciones y correcciones para controles (CN) en cada fold
        temp_CN_df = pd.DataFrame({
            'y_labels': y_test_kf_CN,
            'y_pred': y_pred_CN_test,
            'y_pred_corrected': y_pred_corrected_CN,
            'GAP': gap_CN,
            'GAP_corrected': corrected_gap_CN,
            'ID': id_test_kf_CN
        })
        results_labels_df_CN = pd.concat([results_labels_df_CN, temp_CN_df], ignore_index=True)
        results_per_fold_CN[fold_CN] = temp_CN_df.copy()

        # Almacenar los resultados de predicciones y correcciones para AD
        temp_AD_df = pd.DataFrame({
            'y_labels': y_test_kf_AD,
            'y_pred': y_pred_AD_test,
            'y_pred_corrected': y_pred_corrected_AD,
            'GAP': gap_AD,
            'GAP_corrected': corrected_gap_AD,
            'ID': id_test_kf_AD
        })
        results_labels_df_AD = pd.concat([results_labels_df_AD, temp_AD_df], ignore_index=True)
        results_per_fold_AD[fold_CN] = temp_AD_df.copy()

        # Almacenar los resultados de predicciones y correcciones para FTD
        temp_FTD_df = pd.DataFrame({
            'y_labels': y_test_kf_FTD,
            'y_pred': y_pred_FTD_test,
            'y_pred_corrected': y_pred_corrected_FTD,
            'GAP': gap_FTD,
            'GAP_corrected': corrected_gap_FTD,
            'ID': id_test_kf_FTD
        })
        results_labels_df_FTD = pd.concat([results_labels_df_FTD, temp_FTD_df], ignore_index=True)
        results_per_fold_FTD[fold_CN] = temp_FTD_df.copy()

        # Almacenar los resultados de predicciones y correcciones para MCI
        temp_MCI_df = pd.DataFrame({
            'y_labels': y_test_kf_MCI,
            'y_pred': y_pred_MCI_test,
            'y_pred_corrected': y_pred_corrected_MCI,
            'GAP': gap_MCI,
            'GAP_corrected': corrected_gap_MCI,
            'ID': id_test_kf_MCI
        })
        results_labels_df_MCI = pd.concat([results_labels_df_MCI, temp_MCI_df], ignore_index=True)
        results_per_fold_MCI[fold_CN] = temp_MCI_df.copy()

        # Almacenar los resultados de predicciones y correcciones para PD
        temp_PD_df = pd.DataFrame({
            'y_labels': y_test_kf_PD,
            'y_pred': y_pred_PD_test,
            'y_pred_corrected': y_pred_corrected_PD,
            'GAP': gap_PD,
            'GAP_corrected': corrected_gap_PD,
            'ID': id_test_kf_PD
        })
        results_labels_df_PD = pd.concat([results_labels_df_PD, temp_PD_df], ignore_index=True)
        results_per_fold_PD[fold_CN] = temp_PD_df.copy()

        # Almacenar el modelo
        results['model'].append(model)


# Promediar los valores por ID en results_labels_df_CN
results_avg_CN = results_labels_df_CN.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_AD
results_avg_AD = results_labels_df_AD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_FTD
results_avg_FTD = results_labels_df_FTD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_MCI
results_avg_MCI = results_labels_df_MCI.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

# Promediar los valores por ID en results_labels_df_PD
results_avg_PD = results_labels_df_PD.groupby('ID').agg({
    'y_labels': 'mean',
    'y_pred': 'mean',
    'y_pred_corrected': 'mean',
    'GAP': 'mean',
    'GAP_corrected': 'mean'
}).reset_index()

print("Results average for CN:")
print(results_avg_CN.head())

print("Results average for AD:")
print(results_avg_AD.head())

print("Results average for FTD:")
print(results_avg_FTD.head())

print("Results average for MCI:")
print(results_avg_MCI.head())

print("Results average for PD:")
print(results_avg_PD.head())
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import linregress
import pandas as pd
import numpy as np

def entrenar_modelo(df_concatenado_CN, lista_dfs, n_splits=10, n_iterations=20):
    # Preparar el dataframe de controles
    X_CN = df_concatenado_CN.iloc[:, :-2]  # Features
    y_CN = df_concatenado_CN.iloc[:, -2]   # Labels (Age)
    ID_CN = df_concatenado_CN.iloc[:, -1]  # IDs

    # Inicializar resultados
    results = {'model': []}
    results_labels_df_CN = pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID'])
    results_labels_list = []

    # Crear dataframes para almacenar resultados de pacientes
    for df in lista_dfs:
        results_labels_list.append(pd.DataFrame(columns=['y_labels', 'y_pred', 'y_pred_corrected', 'GAP', 'GAP_corrected', 'ID']))

    # Bucle de iteraciones
    for i in range(n_iterations):
        # Crear validación cruzada para CN
        kf_CN = KFold(n_splits=n_splits, shuffle=True, random_state=i)
        kf_CN_splits = list(kf_CN.split(X_CN, y_CN))

        # Crear validación cruzada para cada dataframe de pacientes
        kf_splits_list = [list(KFold(n_splits=n_splits, shuffle=True, random_state=i).split(df.iloc[:, :-2], df.iloc[:, -2])) for df in lista_dfs]

        for fold in range(n_splits):
            # Obtener índices de entrenamiento y prueba para CN
            train_index_CN, test_index_CN = kf_CN_splits[fold]
            X_train_kf_CN, X_test_kf_CN = X_CN.iloc[train_index_CN], X_CN.iloc[test_index_CN]
            y_train_kf_CN, y_test_kf_CN = y_CN.iloc[train_index_CN], y_CN.iloc[test_index_CN]
            id_test_kf_CN = ID_CN.iloc[test_index_CN]

            # Escalar datos de entrenamiento y prueba para CN
            mean_X_train_kf = X_train_kf_CN.mean()
            std_X_train_kf = X_train_kf_CN.std()
            X_train_kf_CN_scaled = (X_train_kf_CN - mean_X_train_kf) / std_X_train_kf
            X_test_kf_CN_scaled = (X_test_kf_CN - mean_X_train_kf) / std_X_train_kf

            # Entrenar el modelo con CN
            model = Lasso(alpha=0.2, max_iter=10000)
            model.fit(X_train_kf_CN_scaled, y_train_kf_CN)
            
            y_pred_train = model.predict(X_train_kf_CN_scaled)
            gap_train = y_pred_train - y_train_kf_CN

            # Hacer predicciones para el conjunto de prueba de CN
            y_pred_CN_test = model.predict(X_test_kf_CN_scaled)
            gap_CN = y_pred_CN_test - y_test_kf_CN

            # Ajuste de GAP para CN
            slope, intercept, _, _, _ = linregress(y_train_kf_CN, gap_train)
            corrected_gap_CN = gap_CN - (slope * y_test_kf_CN + intercept)
            y_pred_corrected_CN = y_pred_CN_test - (slope * y_test_kf_CN + intercept)

            # Guardar resultados de CN
            temp_CN_df = pd.DataFrame({
                'y_labels': y_test_kf_CN,
                'y_pred': y_pred_CN_test,
                'y_pred_corrected': y_pred_corrected_CN,
                'GAP': gap_CN,
                'GAP_corrected': corrected_gap_CN,
                'ID': id_test_kf_CN
            })
            results_labels_df_CN = pd.concat([results_labels_df_CN, temp_CN_df], ignore_index=True)

            # Procesar cada dataframe de pacientes
            for j, df in enumerate(lista_dfs):
                train_index_pat, test_index_pat = kf_splits_list[j][fold]
                X_train_pat = df.iloc[train_index_pat, :-2]
                X_test_pat = df.iloc[test_index_pat, :-2]
                y_test_pat = df.iloc[test_index_pat, -2]
                id_test_pat = df.iloc[test_index_pat, -1]

                # Escalar usando los parámetros de CN
                X_test_pat_scaled = (X_test_pat - mean_X_train_kf) / std_X_train_kf

                # Predicciones para el grupo de pacientes
                y_pred_pat_test = model.predict(X_test_pat_scaled)
                gap_pat = y_pred_pat_test - y_test_pat

                # Ajuste de GAP para los pacientes
                corrected_gap_pat = gap_pat - (slope * y_test_pat + intercept)
                y_pred_corrected_pat = y_pred_pat_test - (slope * y_test_pat + intercept)

                # Guardar resultados para cada grupo de pacientes
                temp_pat_df = pd.DataFrame({
                    'y_labels': y_test_pat,
                    'y_pred': y_pred_pat_test,
                    'y_pred_corrected': y_pred_corrected_pat,
                    'GAP': gap_pat,
                    'GAP_corrected': corrected_gap_pat,
                    'ID': id_test_pat
                })
                results_labels_list[j] = pd.concat([results_labels_list[j], temp_pat_df], ignore_index=True)

            # Guardar el modelo entrenado
            results['model'].append(model)

    return results_labels_df_CN, results_labels_list, results
```

```python
from sklearn.base import BaseEstimator, RegressorMixin
import numpy as np

class EnsembleWrapper(BaseEstimator, RegressorMixin):
    def __init__(self, models=None):
        # models es una lista de instancias entrenadas de MLPWrapper
        self.models = models if models is not None else []
    
    def fit(self, x_train, y_train):
        # Aquí no es necesario ajustar ya que los modelos están pre-entrenados
        pass

    def predict(self, x_test):
        # Promediar las predicciones de todos los modelos
        predictions = np.array([model.predict(x_test) for model in self.models])
        return np.mean(predictions, axis=0)

    def score(self, x_test, y_test):
        y_pred = self.predict(x_test)
        return np.mean((y_test - y_pred) ** 2)  # MSE como ejemplo
```
