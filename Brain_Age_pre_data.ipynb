```python
# Listado de datasets
datasets = [df_AD_arg, df_AD_col, df_AD_bra, df_AD_ch, df_AD_tur]

# Comprobar si todos los datasets tienen el mismo formato de columnas
columnas_iguales = all(df.columns.equals(datasets[0].columns) for df in datasets)

if columnas_iguales:
    print("Todos los datasets tienen el mismo formato de columnas.")
else:
    print("Los datasets no tienen el mismo formato de columnas.")
```
