```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import pickle


file_path_CN = f'{path}Brain_Age_with_Patient/Data/CN_freq.xlsx'
file_path_AD = f'{path}Brain_Age_with_Patient/Data/AD_freq.xlsx'
file_path_FTD = f'{path}Brain_Age_with_Patient/Data/FTD_freq.xlsx'
file_path_MCI = f'{path}Brain_Age_with_Patient/Data/MCI_freq.xlsx'
file_path_PD = f'{path}Brain_Age_with_Patient/Data/PD_freq.xlsx'

df_CN = pd.read_excel(file_path_CN)
df_AD = pd.read_excel(file_path_AD)
df_FTD = pd.read_excel(file_path_FTD)
df_MCI = pd.read_excel(file_path_MCI)
df_PD = pd.read_excel(file_path_PD)

df_CN_filtrado = df_CN[(df_CN['Age'] >=50) & (df_CN['Age'] <= 90)].reset_index(drop=True)
df_AD_filtrado = df_AD[(df_AD['Age'] >= 50) & (df_AD['Age'] <= 90)].reset_index(drop=True)
df_FTD_filtrado = df_FTD[(df_FTD['Age'] >= 50) & (df_FTD['Age'] <= 90)].reset_index(drop=True)
df_MCI_filtrado = df_MCI[(df_MCI['Age'] >= 50) & (df_MCI['Age'] <= 90)].reset_index(drop=True)
df_PD_filtrado = df_PD[(df_PD['Age'] >= 50) & (df_PD['Age'] <= 90)].reset_index(drop=True)
```

# CN

```python
countries = df_CN_filtrado['Country_ID'].str.split('_').str[0]
unique_countries = countries.unique()
unique_countries_list = unique_countries.tolist()
print(unique_countries_list)
```

```python
# Extraer los códigos de país de la columna 'Country_ID'
df_CN_filtrado['Country_'] = df_CN_filtrado['Country_ID'].str.split('_').str[0]

# Crear un diccionario para mapear los códigos a los nombres completos
country_mapping = {
    'CUB': 'Cuba',
    'ARG': 'Argentina',
    'CHI': 'Chile',
    'TUR': 'Turquia',
    'IRL': 'Irlanda',
    'COL': 'Colombia',
    'CALI': 'Colombia',  # Cali mapeado a Colombia
    'BRA': 'Brasil',
    'GRE': 'Grecia',
    'ITA': 'Italia',
    'UK': 'Reino Unido'
}

# Crear la nueva columna 'Country-unique' usando el mapeo
df_CN_filtrado['Country-unique'] = df_CN_filtrado['Country_'].map(country_mapping)

# Verificar los resultados
print(df_CN_filtrado[['Country_ID', 'Country_', 'Country-unique']])
```

```python
# Mapear los países a regiones
region_mapping = {
    'Cuba': 'Cuba',
    'Turquia': 'Turquia',
    'Irlanda': 'Europa',
    'Grecia': 'Europa',
    'Italia': 'Europa',
    'Reino Unido': 'Europa',
    'Argentina': 'America',
    'Chile': 'America',
    'Brasil': 'America',
    'Colombia': 'America'
}

# Crear la columna 'Region' en el dataframe
df_CN_filtrado['Region'] = df_CN_filtrado['Country-unique'].map(region_mapping)
```

```python
df_CN_filtrado.drop(columns=['Country_'], inplace=True)
```

```python
df_CN_filtrado.to_excel(file_path_CN, index=False)
```
